```python
import cv2
from google.colab.patches import cv2_imshow
```

```python
img1=cv2.imread('img_source1.jpg')
cv2_imshow(img1)
```

```python
img1_gray=cv2.cvtColor(img1,cv2.COLOR_RGB2GRAY)
cv2_imshow(img1_gray)
```

```python
thres, img1_result = cv2.threshold(img1_gray, 160,255,cv2.THRESH_BINARY)
thres, img1_inv = cv2.threshold(img1_gray, 140,255,cv2.THRESH_BINARY_INV)
cv2_imshow(img1_result)
cv2_imshow(img1_inv)
```

```python
#otsu algorithm
a, img_result_otsu=cv2.threshold(img1_gray, 0,255,cv2.THRESH_BINARY+cv2.THRESH_OTSU)
cv2_imshow(img_result_otsu)
```

```python
img2=cv2.imread('img_source2.png')
img2_gray=cv2.cvtColor(img2,cv2.COLOR_RGB2GRAY)
cv2_imshow(img2_gray)
```

```python
img2_result_adapt=cv2.adaptiveThreshold(img2_gray,255,cv2.ADAPTIVE_THRESH_MEAN_C,cv2.THRESH_BINARY,21,5)
cv2_imshow(img2_result_adapt)
```

```python
img2_blur=cv2.GaussianBlur(img2_gray,(5,5),0)
img2_result_blur_adapt=cv2.adaptiveThreshold(img2_blur,255,cv2.ADAPTIVE_THRESH_MEAN_C,cv2.THRESH_BINARY,21,5)
cv2_imshow(img2_result_blur_adapt)
```
